```python
m, k = indices.shape
# Weight matrix.
I = np.arange(0, m).repeat(k)
J = indices.reshape(m * k)
B = distances.reshape(m * k)
```

```python
def upper_triangle_array_diag(matrix):
    result = matrix[np.triu_indices_from(matrix)]
    return result
```

```python
def reconstruct_symmetric_matrix_diag(upper_triangle_array, n):
    result = np.zeros((n, n))
    result[np.triu_indices_from(result)] = upper_triangle_array
    result = result + result.T - np.diag(result.diagonal())
    return result
```
